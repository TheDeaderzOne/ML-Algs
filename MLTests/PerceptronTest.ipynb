```python
import pandas as pd
from sklearn.datasets import make_blobs
from CompletedMLAlgorithms.Perceptron import Perceptron
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

Dataset Generation

```python
df, target = make_blobs(n_samples=10000, centers=2, n_features=2, cluster_std=2.5, random_state=42)
df, testdf, df_target, testdf_target = train_test_split(df, target, test_size=0.25, random_state=42)
```

Fitting & Predicting

```python
model = Perceptron(n_iterations=100)
model.fit(df, df_target)
```

```python
print(accuracy_score(model.predict(testdf), testdf_target))
```
